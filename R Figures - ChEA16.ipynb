```r
#%%R -w 1500 -h 1000

library(ggplot2)
library(ggExtra)
library(WVPlots)

par(mfrow=c(1,2))

pvalueUP <-chea$pvalueUP
pvaldn <- chea$pvaldn

lg10up <- chea$lg10up 
lg10dn <-  chea$lg10dn 

#  <-chea$pvalBFup
# <- chea$pvalBFdn
# chea$pvalFDRup
# chea$pvalFDRdn

## A ### Raw p value
plot_center = ggplot(chea, aes(x=pvalueUP,y=pvaldn)) + 
  geom_point() +
  geom_smooth(method="lm")

# default density
ggMarginal(plot_center, type="histogram")
ScatterHist(chea, "pvalueUP", "pvaldn",
            smoothmethod="lm",
            binwidth_x = 10,
            binwidth_y = 10,
            annot_size=2,
            title="ChEA Raw p values")


# ## B ### log 10 p values
plot_center = ggplot(chea, aes(x=lg10up,y=lg10dn)) + 
  geom_point() +
  geom_smooth(method="lm")

# default density
ggMarginal(plot_center, type="histogram")

ScatterHist(chea, "lg10up", "lg10dn",
            smoothmethod="lm",
            binwidth_x = 10,
            binwidth_y = 10,
            annot_size=2,
            title="ChEA log transformed p values")


## B ### Bonferroni p values
plot_center = ggplot(chea, aes(x=pvalBFup,y=pvalBFdn)) + 
  geom_point() +
  geom_smooth(method="lm")

# default density
ggMarginal(plot_center, type="histogram")

ScatterHist(chea, "pvalBFup", "pvalBFdn",
            smoothmethod="lm",
            binwidth_x = 10,
            binwidth_y = 10,
            annot_size=2,
            title="ChEA Bonferroni p values")


## B ### FDR p values
plot_center = ggplot(chea, aes(x=pvalFDRup,y=pvalFDRdn)) + 
  geom_point() +
  geom_smooth(method="lm")

# default density
ggMarginal(plot_center, type="histogram")

ScatterHist(chea, "pvalFDRup", "pvalFDRdn",
            smoothmethod="lm",
            binwidth_x = 10,
            binwidth_y = 10,
            annot_size=2,
            title="ChEA FDR p values")
```
